```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data/spider.csv")
```

```python
my_data = data.iloc[6400:,:]
my_data
```

```python
my_data["HAVING"].plot()
```

```python
my_data["JOIN"].plot()
```

```python
my_data["GROUP BY"].plot()
```

```python
my_data.loc[:,'Total'] = my_data[["HAVING", "JOIN", "GROUP BY", "ORDER BY"]].copy().sum(axis=1)
my_data
```

```python
my_data.sort_values(by=['Total'], ascending=False, inplace=True)
my_data
```
