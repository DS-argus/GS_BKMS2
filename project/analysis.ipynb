```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("spider_250.csv")
data.head()
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

data["HAVING"].plot(ax=axes[0], title="HAVING")
data["ORDER BY"].plot(ax=axes[1], title="ORDER BY")
data["GROUP BY"].plot(ax=axes[2], title="GROUP BY")
data["JOIN"].plot(ax=axes[3], title="JOIN")
data["SELECT"].plot(ax=axes[4], title="SELECT")
```

```python
print(len(data['db_id'].unique()))
plt.figure(figsize=(25, 5))
ordered_categories = data['db_id'].value_counts().index
sns.countplot(x='db_id', data=data, order=ordered_categories)
plt.xticks(rotation=90);
```
